```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
df.info()
```

Q1: Jumlah baris dan kolom dataset Titanic serta nama kolom

```python
df.shape
df.columns
```

Dataset Titanic memiliki 891 baris dan 12 kolom.
Kolom yang tersedia adalah PassengerId, Survived, Pclass, Name, Sex, Age, SibSp, Parch, Ticket, Fare, Cabin, dan Embarked.

Q2: Kolom yang memiliki missing values

```python
df.isnull().sum()
df.isnull().mean()*100
```

Kolom yang memiliki missing values adalah Age, Cabin, dan Embarked.
Kolom Cabin memiliki jumlah missing paling banyak.
Persentase missing terbesar terdapat pada kolom Cabin, diikuti oleh Age dan Embarked.

Q3: Persentase penumpang yang selamat

```python
df['Survived'].mean()*100
```

Persentase penumpang yang selamat adalah 38,3%

Q4: Penumpang wanita kelas 1

```python
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]

jumlah = wk1.shape[0]
rata_usia = wk1['Age'].mean()

jumlah, rata_usia
```

Jumlah penumpang wanita kelas 1 adalah 94 orang.
Rata-rata usia penumpang wanita kelas 1 adalah sekitar 34 tahun.
Hal ini menunjukkan bahwa mayoritas penumpang wanita kelas 1 berada pada usia dewasa.

Q5: Tingkat keselamatan per kelas penumpang

```python
df.groupby('Pclass')['Survived'].mean()
```

Tingkat keselamatan tertinggi terdapat pada penumpang kelas 1, diikuti kelas 2, dan paling rendah kelas 3.
Hal ini menunjukkan bahwa penumpang kelas atas memiliki peluang lebih besar untuk selamat dibandingkan kelas bawah.

Q6: Analisis tambahan berdasarkan jenis kelamin

```python
df.groupby('Sex')['Survived'].mean()
```

Penumpang perempuan memiliki tingkat keselamatan yang jauh lebih tinggi dibandingkan laki-laki.
Hal ini menunjukkan adanya prioritas penyelamatan terhadap perempuan saat kejadian Titanic.
